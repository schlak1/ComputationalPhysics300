```python
#import the needed librarys/functions
import matplotlib.pyplot as plt
import numpy as np
#establish variables and arrays
N = 1000
points = np.zeros([N,N],int)
xpoint = np.linspace(-2,2,N)
ypoint = np.linspace(-2,2,N)
#function to determine valids points
for a, x in enumerate(xpoint):
    for b, y in enumerate(ypoint):
        z = 0
        c = complex(x,y)
        for i in range(N):
            z=z*z+c
            if abs(z) > 2:
                points[b,a] = 1
                break

# Make the plot
plt.imshow(points, origin="lower")
plt.gray()
plt.show()
```
