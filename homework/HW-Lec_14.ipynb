```python
# Brownian Motion
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=[18, 12])

# initialization
L1 = 101
L2 = 101
N = 1000

position = np.zeros([N,2])
position[0,0] = int(L1/2)
position[0,1] = int(L2/2)

direction = np.array([[0,1],[0,-1],[1,0],[-1,0]]) #up, down, right, left

#np.random.seed(2) 

# random walk
for i in range(1,N):    
    while True:
        move = np.random.randint(len(direction))
        position_next = position[i-1,:] + direction[move]
        #boundary condition
        if 0<position_next[0]+1<L1+2 and 1<position_next[1]+1<L2+2:
            break
    
    #update the positions for step i.
    position[i,0] = position_next[0]
    position[i,1] = position_next[1]
    #position[i,:] = position_next

# plot the trajectory
ax = plt.axes()
ax.set_facecolor('silver')
plt.xlim(0,L1)
plt.ylim(0,L2)
plt.title('Brownian motion')
plt.scatter(position[:,0], position[:,1], c=range(N), s=60, alpha=0.8)
plt.text(position[0,0], position[0,1], 'start', fontsize=24, color='w')
plt.text(position[-1,0], position[-1,1], 'end', fontsize=24)

cbar = plt.colorbar()
cbar.set_label('Itertion number')
plt.show()
```
